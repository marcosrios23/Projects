```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_csv('ca_he.csv')
```

```python
raw_data
```

```python
data = raw_data.transpose()
data.drop(['Series Name','Series Code','Country Code'], inplace= True)
data = data.iloc[:,:7]
data = data.rename(columns = data.loc['Country Name'])
data.drop('Country Name', inplace = True)
data
```

```python
data = data.rename(index = lambda a: a[:4])
data = data.replace('..', np.NaN)
data
```

```python
data.dtypes
```

```python
data = data.astype('float64')
data = data.fillna(method='ffill')
```

```python
plt.figure(figsize=[15,5])
for i in range(7):
    plt.plot(data.iloc[:,i], label = data.columns[i])
plt.legend()
plt.show()
```
